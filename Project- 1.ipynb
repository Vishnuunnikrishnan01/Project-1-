### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_counts =df['Team'].value_counts()
team_counts
```

```python
total_employees = len(df)
total_employees
```

```python
percentage = (team_counts/total_employees)*100
percentage
```

### 2.Segregate the employees w.r.t different positions.

```python
pos_grp = df.groupby("Position")["Name"].apply((list))
for position,names in pos_grp.items():
    print(f"Employees in positions {position}:")
    for name in names:
        print(name)
```

### 3.Find from which age group most of the employees belong to.

```python
age_grp= df['Age'].value_counts()
age_grp
```

```python
most_common_ageGrp =age_grp.idxmax()
print("Age group most of the employees belong to is :",most_common_ageGrp)
```

### 4.Find out under which team and position, spending in terms of salary is high.

```python
spending_high_salary = df.groupby(['Team','Position'])['Salary'].sum()
spending_high_salary.idxmax()
```

### 5.Find if there is any correlation between age and salary , represent it visually.

```python
import matplotlib.pyplot as plt
```

```python
correlation =df['Age'].corr(df['Salary'])
print(f'Correlation between Age and Salary =',correlation)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['Salary'],color='magenta')
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
```
